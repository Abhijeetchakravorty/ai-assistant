```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
data_explore = data.copy()
actual = data_explore["actualPrice"].str.split("₹", n = 1, expand = True)
discounted = data_explore.discountedPrice.str.split("₹", n = 1, expand = True)
actualD = actual[1].str.replace(",", "").astype(float)
actualD = np.array(actualD)
only_data = discounted.to_numpy()
discountedArr = []
for i in range(len(only_data)):
    for j in range(len(only_data[i])):
        discountedArr.append(only_data[i][j])
for k in range(len(discountedArr)):
    discountedArr[k] = discountedArr[k].replace(",", "")
    discountedArr[k] = int(discountedArr[k])
fig, ax = plt.subplots(figsize=(15, 10))
sns.barplot(actualD, dat["name"])
plt.xscale('log')
```
